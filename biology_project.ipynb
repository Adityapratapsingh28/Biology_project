```python
# Changing gender values to 1 (male) and 0 (female)
df['Smoking History'] = df['Smoking History'].map({'Current ': 1, 'Never': 0 , ' Former ': 1})
df['Smoking History'] = df['Smoking History'].fillna(1)
```

```python
df['Physical Activity'] = df['Physical Activity'].map({'High': 1, 'Low': 0 , ' Moderate': 1})
df['Physical Activity'] = df['Physical Activity'].fillna(0)
```

```python
df['Alcohol Consumption'] = df['Alcohol Consumption'].map({'Regualr': 1, 'Never': 0 , ' Occasional': 1})
df['Alcohol Consumption'] = df['Alcohol Consumption'].fillna(1)
```
